# **Mosquito Indicator em Seoul, Cor√©ia do Sul**
[Link para o Dataset no Kaggle](https://www.kaggle.com/datasets/kukuroo3/mosquito-indicator-in-seoul-korea/data)


---

#### **Dados**
Este conjunto de dados trata do **√çndice de Mosquitos** e das **condi√ß√µes clim√°ticas (temperatura e precipita√ß√£o)** em **Seul, Coreia do Sul**.  
O **√çndice de Mosquitos** representa o n√∫mero de mosquitos por uma determinada √°rea espec√≠fica.

---

#### **Conte√∫do**
Este conjunto de dados possui **seis colunas**:

- `date` ‚Äì data da medi√ß√£o  
- `mosquito_Indica` ‚Äì √≠ndice de mosquitos (quantidade por √°rea espec√≠fica)  
- `rain(mm)` ‚Äì precipita√ß√£o (chuva) em mil√≠metros  
- `mean_T(‚ÑÉ)` ‚Äì temperatura m√©dia (¬∞C)  
- `min_T(‚ÑÉ)` ‚Äì temperatura m√≠nima (¬∞C)  
- `max_T(‚ÑÉ)` ‚Äì temperatura m√°xima (¬∞C)

---

#### **Per√≠odo da coleta**
Os dados foram medidos **diariamente**, entre **maio de 2016** e **dezembro de 2019**, em uma **√°rea espec√≠fica de Seul**.

---

#### **Fonte:**
Os dados foram disponibilizados neste site oficial:  
üîó [https://news.seoul.go.kr/welfare/mosquito](https://news.seoul.go.kr/welfare/mosquito)


```python
import pandas as pd
import numpy as np
from google.colab import files
uploaded = files.upload()

train_dataset = pd.read_csv('mosquito_Indicator.csv')
print(train_dataset['mosquito_Indicator'])

X = train_dataset['mosquito_Indicator']
Y = train_dataset['rain(mm)']

mediaX = np.mean(X)
mediaY = np.mean(Y)

stdX = np.std(train_dataset['mosquito_Indicator'], ddof=1)
stdY = np.std(train_dataset['rain(mm)'], ddof=1)
print('')
print("M√©dia de Mosquitos : ", mediaX)
print("M√©dia de precipita√ß√£o : ", mediaY)
print("Desvio Padr√£o dos Mosquitos : ", stdX)
print("Desvio Padr√£o das precipita√ß√µes : ", stdY)

deviationX = X - mediaX
deviationY = Y - mediaY
sumCovariance = np.sum(deviationX * deviationY)
print('')
print("Soma da Covari√¢ncia : ", sumCovariance)

N = train_dataset.shape[0]

pearsonCorrelation = sumCovariance / ((N - 1) * stdX * stdY)
print('')
print("Correla√ß√£o de Pearson : ", pearsonCorrelation)
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()

train_dataset = pd.read_csv('mosquito_Indicator.csv')

X = train_dataset['mosquito_Indicator']
Y = train_dataset['rain(mm)']

slope, intercept, r_value, p_value, std_err = stats.linregress(X, Y)

print("(Beta) Inclina√ß√£o : ", slope)
print("(Alpha) Intercept : ", intercept)
print("Coeficiente Correla√ß√£o : ", r_value)
print("P_Value : ", p_value)
print("Erro do Desvio Padr√£o : ", std_err)

plt.title('Mosquito Indicator')
plt.xlabel('N√∫mero de Mosquitos em uma √°rea espec√≠fica')
plt.ylabel('Precipita√ß√£o (mm)')
plt.plot(X, Y, 'o', label='original_data')
plt.plot(X, intercept + slope * X, 'r', label='fitted_line')
plt.legend()

plt.show()
```
